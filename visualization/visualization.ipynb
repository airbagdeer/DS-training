```python
import matplotlib as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
titanic_dataframe = pd.read_csv('titanic.csv')
```

```python
titanic_dataframe.columns
```

```python
sns.countplot(data=titanic_dataframe, x='sex', hue='sex', palette='Set1')
```

```python
sns.boxplot(data=titanic_dataframe, x='pclass', y='age', palette='Set1', hue='sex')
```

```python
top_family_casualties = titanic_dataframe['name'].map(lambda x: x.split(',')[0]).value_counts().sort_values(ascending=False)[:5]
sns.barplot(x=top_family_casualties.index, y=top_family_casualties.values, palette='Set1', hue=top_family_casualties.index)
```

```python
sns.boxplot(titanic_dataframe, x='pclass', y='age', hue='survived', palette='Set1')
```

```python
sns.countplot(data=titanic_dataframe, x='pclass', hue='survived', palette='Set1')
```
